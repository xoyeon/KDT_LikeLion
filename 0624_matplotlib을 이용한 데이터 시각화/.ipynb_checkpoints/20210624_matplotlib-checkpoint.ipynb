# plot, bar 그래프 그리기

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

x=range(0,100)
y=[v*v for v in x]

ax1.plot(x,y)
ax2.bar(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 2 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)

fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(x, sin_y, 'b--')
ax2.plot(x, cos_y, 'r--')
plt.show()
```

#### (실습) sin, cos, tan, tanh
 * 2 x 2 열의 그래프로 그려보기

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 2 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)
tan_y = np.tan(x)
tanh_y = np.tanh(x)

fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)


ax1.plot(x, sin_y, 'b--')
ax2.plot(x, cos_y, 'r--')
ax3.plot(x, tan_y, 'b--')
ax4.plot(x, tanh_y, 'r--')

plt.show()
```

```python
x = np.arange(10)
plt.plot(x, 1*x, label='y = %ix' % 1)
plt.plot(x, 2*x, label='y = %ix' % 2)
plt.legend()
plt.show()
```

# 범례 위치 변경과 subplots() 함수

```python
plt.figure(figsize=(15,10))
plt.subplot(3,3,1)
plt.plot(x, 1*x, label='y = %ix' % 1)
plt.legend(loc='upper left')    # 2
plt.subplot(3,3,2)
plt.plot(x, 1*x, label='y = %ix' % 1)
plt.legend(loc='upper center')  # 9
plt.subplot(3,3,3)
plt.plot(x, 1*x, label='y = %ix' % 1)
plt.legend(loc='upper right')  # 1
plt.subplot(3,3,4)
plt.plot(x, 1*x, label='y = %ix' % 1)
plt.legend(loc='center left')  # 6
plt.subplot(3,3,5)
plt.plot(x, 1*x, label='y = %ix' % 1)
plt.legend(loc='center')  # 10
plt.subplot(3,3,6)
plt.plot(x, 1*x, label='y = %ix' % 1)
plt.legend(loc='center right')  # 5
plt.subplot(3,3,7)
plt.plot(x, 1*x, label='y = %ix' % 1)
plt.legend(loc='lower left')  # 3
plt.subplot(3,3,8)
plt.plot(x, 1*x, label='y = %ix' % 1)
plt.legend(loc='lower center')  # 8
plt.subplot(3,3,9)
plt.plot(x, 1*x, label='y = %ix' % 1)
plt.legend(loc='lower right')  # 4
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10, size=10)
print(len(x), type(x))
y = np.random.randint(80,100, size=10)
print(len(y), type(y))

colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()   # colormap 을 포함할 수 있다.

plt.show()
```
