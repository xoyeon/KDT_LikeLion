## select 함수 활용

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
m_url = 'https://movie.naver.com/movie/running/current.nhn#'
page = urlopen(m_url)
soup = BeautifulSoup(page, "html.parser")
movies = soup.select('#content > div.article > div:nth-child(1) > div.lst_wrap > ul > li')
print('영화제목 ','평점','참여','애매율\t','감독')
for movie in movies:
    title = movie.select_one(' dl > dt > a').text
    star = movie.select_one('li > dl > dd.star > dl.info_star > dd > div > a > span.num').text
    people = movie.select_one('li > dl > dd.star > dl.info_star > dd > div > a > span.num2 > em').text
    rate = movie.select_one('li > dl > dd.star > dl.info_exp > dd > div > span.num')
    king = movie.select_one('li > dl > dd:nth-child(3)> dl > dd:nth-child(4) > span > a').text
    try:
        print(title, star, people, rate, king)
    except Exception as e:
        continue
```

### select 함수 간단히

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
m_url = 'https://movie.naver.com/movie/running/current.nhn#'
page = urlopen(m_url)
soup = BeautifulSoup(page, "html.parser")
movies = soup.select('#content > div.article > div:nth-child(1) > div.lst_wrap > ul > li')
print('영화제목\t ','평점\t','참여\t','애매율\t\t','감독')
for movie in movies:
    title = movie.select_one(' dl > dt > a').text
    star = movie.select_one('span.num').text
    people = movie.select_one('em').text
    rate = movie.select_one(' div > span.num')
    king = movie.select_one('dd:nth-child(4) > span > a').text
    try:
        print(title+'\t', star+'\t', people+'\t', rate.text+'\t', king)
    except Exception as e:
        continue
        
# 원래는 text만 추출하는거라 에러가 안나는데 rate부분은 예매율이라서 숫자로 추출이 되더라구요(숫자로추출이 되어서 타입에러가남)!!
# 그래서 예외처리를 해줘서 에러메세지가 안나게 해준겁니당!!
```

### 조건문 사용하여 작성

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
m_url = 'https://movie.naver.com/movie/running/current.nhn#'
page = urlopen(m_url)
soup = BeautifulSoup(page, "html.parser")
movies = soup.select('#content > div.article > div:nth-child(1) > div.lst_wrap > ul > li')
print('영화제목\t ','평점\t','참여\t','애매율\t\t','감독\n')
for movie in movies:
    title = movie.select_one(' dl > dt > a').text
    star = movie.select_one('span.num').text
    people = movie.select_one('em').text
    rate = movie.select_one(' div > span.num')
    king = movie.select_one('dd:nth-child(4) > span > a').text
    if rate is not None:
        a = rate.text
        print(title + '\t', star + '\t', people + '\t', a + '\t', king + '\n')
    else:
        a = '0'
        print(title + '\t', star + '\t', people + '\t', a + '\t', king + '\n')
```

## find 함수 활용하여 작성 해보기

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = "https://movie.naver.com/movie/running/current.nhn"
page = urlopen(url)
soup = BeautifulSoup(page, 'lxml')
```

```python
soup_ul_li = soup.find("ul", class_="lst_detail_t1").find_all("li")
len(soup_ul_li)
```

### 제목

```python
soup_ul_li[129].find("dt", class_="tit").a.text
```

### 평점

```python
soup_ul_li[126].find("span", class_="num").text
```

### 예매율

```python
soup_ul_li[3].find("dl", class_="info_exp").span.text
```

### 참여인원

```python
soup_ul_li[0].find("em").text
```

### 감독

```python
soup_ul_li[0].find("dl", class_="info_txt1").findAll("dd")[1].text.strip()
```

```python
soup_ul_li = soup.find("ul", class_="lst_detail_t1").find_all("li")

all_title = []
all_score = []
all_people = []
all_rate = []
all_director = []

for one in soup_ul_li:
    one_title = one.find("dt", class_="tit").a.text  # 제목
    one_score = one.find("span", class_="num").text  # 평점
    one_people = one.find("em").text                 # 참여 인원
    
    tmp_rate = one.find("dl", class_="info-exp")     # 예매율
    if tmp_rate is not None:
        one_rate = one.find("dl", class_="info_exp").span.text
    else:
        one_rate= '0'
    
    # 감독
    one_director = one.find("dl", class_="info_txt1").findAll("dd")[1].text
    one_director = one_director.translate(  { ord('\n'):"", ord('\r'):"", ord('\t'):""} )
        

    
    all_title.append(one_title)
    all_score.append(one_score)
    all_people.append(one_people)
    all_rate.append(one_rate)
    all_director.append(one_director)
    
print(len(all_title), len(all_score), len(all_people), len(all_rate), len(all_director))

print(all_title)
print(all_score)
print(all_people)
print(all_director)
```
